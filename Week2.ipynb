Cleaning Individual Columns at a time for better results 


1.Item Column

```python
df["Item"].value_counts()
```

```python
df['Item'].replace('UNKNOWN', np.nan, inplace=True)
df["Item"].value_counts()
```

```python
error = df[df['Item'] == 'ERROR'].index
df.drop(error, inplace=True)
df["Item"].value_counts()
```

```python
item_mode= df['Item'].mode()[0]
df['Item'].fillna(item_mode, inplace=True)
```

```python
df['Item'].isnull().sum()
```

2. Quantity Column

```python
df['Quantity'].value_counts()
```

```python
df['Quantity'].replace('UNKNOWN', np.nan, inplace=True)
```

```python
error_quantity = df[df['Quantity'] == 'ERROR'].index
df.drop(error_quantity, inplace=True)
```

```python
df['Quantity'].value_counts()
```

```python
df['Quantity']=df['Quantity'].astype('Int64')
```

```python
df['Quantity'].dtypes
```

```python
mean_quantity = df['Quantity'].mean()
df['Quantity'].fillna(round(mean_quantity), inplace=True)
```

```python
df["Quantity"].isnull().sum()
```

```python
df['Quantity'].value_counts()
```

3. Price Per Unit

```python
df['Price Per Unit'].value_counts()
```

```python
df['Price Per Unit'].replace('UNKNOWN', np.nan, inplace=True)
error_price = df[df['Price Per Unit'] == 'ERROR'].index
df.drop(error_price, inplace=True)
```

```python
df['Price Per Unit']=df['Price Per Unit'].astype('float')
df['Price Per Unit'].dtypes
```

```python
mean_price = round(df['Price Per Unit'].mean(), 2)
df['Price Per Unit'].fillna(mean_price, inplace=True)
```

```python
df['Price Per Unit'].value_counts()
```

```python
df['Price Per Unit']= round(df['Price Per Unit'],2)
```

4. Total Spent

```python
df['Total Spent'].value_counts()
```

```python
df['Total Spent'].replace('UNKNOWN', np.nan, inplace=True)
error_total = df[df['Total Spent'] == 'ERROR'].index
df.drop(error_total, inplace=True)
df['Total Spent']=df['Total Spent'].astype('float')
df['Total Spent'].dtypes
```

```python
mean_total = df['Total Spent'].mean()
df['Total Spent'].fillna(mean_total, inplace=True)
```

```python
df['Total Spent']= round(df['Total Spent'],1)
df['Total Spent'].value_counts()
```

```python
df['Total Spent'].isnull().sum()
```

5.Payment Method

```python
df['Payment Method'].value_counts()
```

```python
df['Payment Method'].replace('UNKNOWN', np.nan, inplace=True)
error_payment = df[df['Payment Method'] == 'ERROR'].index
df.drop(error_payment, inplace=True)
df['Payment Method'].value_counts()
```

```python
mode_payment = df['Payment Method'].mode()[0]
df['Payment Method'].fillna(mode_payment, inplace=True)
```

6. Location

```python
df['Location'].value_counts()
```

```python
df['Location'].replace('UNKNOWN', np.nan, inplace=True)
error_location = df[df['Location'] == 'ERROR'].index
df.drop(error_location, inplace=True)
df['Location'].value_counts()
```

```python
mode_location = df['Location'].mode()[0]
df['Location'].fillna(mode_location, inplace=True)
df['Location'].value_counts()
```

7. Transaction Date

```python
df['Transaction Date'].dtypes
```

```python
df['Transaction Date'].value_counts()
```

```python
df['Transaction Date'].replace('UNKNOWN', np.nan, inplace=True)
error_date = df[df['Transaction Date'] == 'ERROR'].index
df.drop(error_date, inplace=True)
```

```python
df['Transaction Date'].isnull().sum()
```

```python
df['Transaction Date']=pd.to_datetime(df['Transaction Date'], errors='coerce')
df['Transaction Date']=df['Transaction Date'].dt.date
df['Transaction Date'].dtypes
```

```python
df['Transaction Date'].isnull().sum()
```

```python
most_frequent_date = df['Transaction Date'].mode()[0]
df['Transaction Date'].fillna(most_frequent_date, inplace=True)
df['Transaction Date'].isnull().sum()
```
